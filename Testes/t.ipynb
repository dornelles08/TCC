```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
demanda = [{"Preço (R$/copo)": 1.50, "Quantidade (copos/dia)": 500}, {"Preço (R$/copo)": 3.00, "Quantidade (copos/dia)": 400},
         {"Preço (R$/copo)": 4.50, "Quantidade (copos/dia)": 300}, {"Preço (R$/copo)": 6.00, "Quantidade (copos/dia)": 200}, 
         {"Preço (R$/copo)": 7.50, "Quantidade (copos/dia)": 100}]


demanda = pd.DataFrame(demanda)
demanda.head()
```

```python
plt.rc('legend', fontsize=15)
plt.rcParams['font.size'] = '15'

plt.figure(figsize=(10, 5))
sns.lineplot(data=demanda, x="Preço (R$/copo)", y="Quantidade (copos/dia)")
plt.xlabel('Quantidade (xícaras/dia)', fontsize=20)
plt.ylabel('Preço (R$/xícara)', fontsize=20)
plt.title('Curva de Demanda', fontsize=20)
plt.legend()
```

```python
supply = [{"Preço (R$/copo)": 1.50, "Quantidade (copos/dia)": 100}, {"Preço (R$/copo)": 3.00, "Quantidade (copos/dia)": 200},
         {"Preço (R$/copo)": 4.50, "Quantidade (copos/dia)": 300}, {"Preço (R$/copo)": 6.00, "Quantidade (copos/dia)": 400}, 
         {"Preço (R$/copo)": 7.50, "Quantidade (copos/dia)": 600}]


supply = pd.DataFrame(supply)
supply.head()
```

```python
plt.rc('legend', fontsize=15)
plt.rcParams['font.size'] = '15'

plt.figure(figsize=(10, 5))
sns.lineplot(data=supply, y="Preço (R$/copo)", x="Quantidade (copos/dia)")
sns.lineplot(data=demanda, y="Preço (R$/copo)", x="Quantidade (copos/dia)")
plt.xlabel('Quantidade (xícaras/dia)', fontsize=20)
plt.ylabel('Preço (R$/xícara)', fontsize=20)
plt.title('Curva de Equilíbrio', fontsize=20)
plt.legend()
```
